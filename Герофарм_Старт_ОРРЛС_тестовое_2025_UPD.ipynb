## **Задание №2.**  

Используя подготовленный на прошлом шаге датасет, постройте графики, которые будут отражать степень сходства (или различия) GPX-007 с молекулами из патентной заявки.  
Вы можете выбрать те варианты визуализации данных, которые сочтете наиболее подходящими. Не забудьте сохранить код для построения графиков в ячейке ниже, чтобы на этапе проверки изображения были воспроизведены корректно.

```python
#ГРАФИК 1: Сравнение дллин боковых радикалов 

# Загружаем таблицу с данными о молекулах
df = pd.read_csv('patent_molecules_table.csv')

# Конвертируем активности в числа (заменяем 'ND' на NaN)
for col in ['abs_glp1r', 'abs_gipr', 'abs_gcgr']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

fig1, ax1 = plt.subplots(figsize=(10, 6))

# GPX-007 выделяем зелёным, остальные - оранжевым
colors = ['green' if x == 'GPX-007' else 'orange' for x in df['id']]

ax1.bar(range(len(df)), df['C_count_tail'], color=colors, alpha=0.7, 
        edgecolor='grey', linewidth=1.5)
ax1.set_xticks(range(len(df)))
ax1.set_yticks(range(21))
ax1.set_xticklabels(df['id'], rotation=45, ha='right', fontsize=10)
ax1.set_ylabel('C_count_tail (carbon atoms)', fontsize=12, fontweight='bold')
ax1.set_xlabel('Compound ID', fontsize=12, fontweight='bold')
ax1.set_title('Comparison of Fatty Acid Chain Length', fontsize=14, fontweight='bold')
ax1.grid(axis='y', alpha=0.3, linestyle='--')
ax1.axhline(y=20, color='gray', linestyle='--', alpha=0.5, label='C20 (GPX-007)')

# Легенда
legend_elements = [Patch(facecolor='green', alpha=0.7, label='GPX-007'),
                   Patch(facecolor='orange', alpha=0.7, label='Patent Compounds')]
ax1.legend(handles=legend_elements, loc='lower right', fontsize=10)

plt.tight_layout()
plt.show()
plt.close()
```

```python
fig4, axes = plt.subplots(1, 2, figsize=(15, 6))

# ГРАФИК A: Количество непротеиногенных аминокислот
ax = axes[0]
colors_aa = ['coral' if x == 'GPX-007' else 'lightgreen' for x in df['id']]
bars = ax.bar(range(len(df)), df['unnatural_aa_count'], 
             color=colors_aa, alpha=0.8, edgecolor='black', linewidth=1.5)

ax.set_xticks(range(len(df)))
ax.set_yticks(range(max(df['unnatural_aa_count'])+1))
ax.set_xticklabels(df['id'], rotation=45, ha='right', fontsize=10)
ax.set_ylabel('Количество аминокислот', fontsize=12, fontweight='bold')
ax.set_title('A. Количество неприродных аминокислот', 
            fontsize=12, fontweight='bold')
ax.grid(axis='y', alpha=0.3, linestyle='--')

for i, (bar, val) in enumerate(zip(bars, df['unnatural_aa_count'])):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.05, 
           str(val), ha='center', va='bottom', fontweight='bold', fontsize=10)

# График B: Распределение branch site, то есть у скольки молекул на каких сайтах ацильный радикал
ax = axes[1]
branch_sites = df['branch_site'].value_counts()
bars = ax.bar(branch_sites.index, branch_sites.values, 
             color=['coral', 'lightgreen'], alpha=0.8, 
             edgecolor='black', linewidth=1.5)

ax.set_ylabel('Частота', fontsize=12, fontweight='bold')
ax.set_xlabel('Позиция присоединения', fontsize=12, fontweight='bold')
ax.set_title('B. Распределение Branch Site', 
            fontsize=12, fontweight='bold')
ax.grid(axis='y', alpha=0.3, linestyle='--')

for bar, val in zip(bars, branch_sites.values):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1, 
           f'n={val}', ha='center', va='bottom', fontweight='bold', fontsize=11)

plt.tight_layout()
plt.show()
plt.close()
```

Ниже я решил сделать отдельные графики для значений EC50, потому что решил, что Compound 18 может частично отражать то, какой активностью будет обладать исследуемое вещество GPX-007 из-за сходства ацильного радикала

```python
# ГРАФИК: Тепловая карта значений EC50 
# (чем меньше значение, тем при меньшей концентрации вещества достигается половина активность половины рецепторов на клетке)
# Отбираем только соединения с известными данными об активности
df_activity = df[df['abs_glp1r'].notna()].copy()
activity_data = df_activity[['abs_glp1r', 'abs_gipr', 'abs_gcgr']].values

# Логарифмическая трансформация
activity_data_log = np.log10(activity_data + 1e-10)

fig2, ax2 = plt.subplots(figsize=(9, 7))

# Создаем heatmap
im = ax2.imshow(activity_data_log, cmap='RdYlGn_r', aspect='auto',
                vmin=-3, vmax=2)

# Настройка осей
ax2.set_xticks(range(3))
ax2.set_xticklabels(['GLP-1R', 'GIPR', 'GCGR'], fontsize=12, fontweight='bold')
ax2.set_yticks(range(len(df_activity)))
ax2.set_yticklabels(df_activity['id'].values, fontsize=11)

ax2.set_title('Значение EC50 log10, nM)',
              fontsize=13, fontweight='bold', pad=20)

# Добавляем значения в ячейки
for i in range(len(df_activity)):
    for j in range(3):
        value = activity_data[i, j]
        if value < 0.01:
            text_str = f'{value:.5f}'
        elif value < 1:
            text_str = f'{value:.3f}'
        else:
            text_str = f'{value:.2f}'
        
        text_color = 'black'
        ax2.text(j, i, text_str, ha="center", va="center",
                color=text_color, fontsize=9, fontweight='bold')

# Colorbar
cbar = plt.colorbar(im, ax=ax2)
cbar.set_label('log10(EC50, nM)', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
plt.close()
```

```python
# 3 графика, которые показывают отношение значений EC50 для разных рецепторов
# Отбираем только соединения с ПОЛНЫМИ данными об активности
df_3d = df[df['abs_glp1r'].notna() & df['abs_gipr'].notna() & df['abs_gcgr'].notna()].copy()

def get_color_and_size(compound_id):
    """Возвращает цвет и размер маркера для каждого соединения."""
    if compound_id == 'Retatrutide':
        return 'orange', 350, '^'
    else:
        return 'green', 250, 'o'


def add_reference_lines(ax, x_name, y_name, x_ref=None, y_ref=None):
    """Добавляет референсные линии на график."""
    if x_ref:
        ax.axvline(x=x_ref, color='gray', linestyle='--', linewidth=1.5, alpha=0.5)
        ax.text(x_ref, ax.get_ylim()[1]*0.95, f'{x_ref}', fontsize=8, alpha=0.7)
    
    if y_ref:
        ax.axhline(y=y_ref, color='gray', linestyle='--', linewidth=1.5, alpha=0.5)
        ax.text(ax.get_xlim()[1]*0.95, y_ref, f'{y_ref}', fontsize=8, alpha=0.7)

fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# ГРАФИК 1: GLP-1R vs GIPR 
ax = axes[0]

for idx, row in df_3d.iterrows():
    color, size, marker = get_color_and_size(row['id'])
    
    ax.scatter(row['abs_glp1r'], row['abs_gipr'],
              c=color, s=size, alpha=0.8,
              edgecolors='black', linewidth=2, marker=marker,
              label=row['id'] if idx == 0 else "")
    
    # Подписи соединений
    ax.annotate(row['id'], 
               (row['abs_glp1r'], row['abs_gipr']),
               xytext=(10, -8), textcoords='offset points', 
               fontsize=9, fontweight='bold')

ax.set_xscale('log')
ax.set_yscale('log')

ax.set_xlabel('GLP-1R EC50, nM)', 
             fontsize=12, fontweight='bold')
ax.set_ylabel('GIPR EC50, nM)', 
             fontsize=12, fontweight='bold')

ax.set_title('ГРАФИК 1: GLP-1R × GIPR', 
            fontsize=13, fontweight='bold', pad=15)

ax.grid(True, alpha=0.3, which='both', linestyle=':')

# Референсные линии (1 nM)
add_reference_lines(ax, 'GLP-1R', 'GIPR', x_ref=1, y_ref=1)

# Легенда
legend_elements = [
    Patch(facecolor='orange', alpha=0.8, label='Retatrutide ▲'),
    Patch(facecolor='green', alpha=0.8, label='Patent Compounds ●')
]
ax.legend(handles=legend_elements, loc='upper left', fontsize=10, framealpha=0.9)



#  ГРАФИК 2: GLP-1R vs GCGR

ax = axes[1]

for idx, row in df_3d.iterrows():
    color, size, marker = get_color_and_size(row['id'])
    
    ax.scatter(row['abs_glp1r'], row['abs_gcgr'],
              c=color, s=size, alpha=0.8,
              edgecolors='black', linewidth=2, marker=marker)
    
    ax.annotate(row['id'], 
               (row['abs_glp1r'], row['abs_gcgr']),
               xytext=(9, -6), textcoords='offset points', 
               fontsize=9, fontweight='bold')

ax.set_xscale('log')
ax.set_yscale('log')

ax.set_xlabel('GLP-1R EC50, nM', 
             fontsize=12, fontweight='bold')
ax.set_ylabel('GCGR EC50, nM', 
             fontsize=12, fontweight='bold')

ax.set_title('ГРАФИК 2: GLP-1R × GCGR', 
            fontsize=13, fontweight='bold', pad=15)

ax.grid(True, alpha=0.3, which='both', linestyle=':')


add_reference_lines(ax, 'GLP-1R', 'GCGR', x_ref=1, y_ref=1)

legend_elements = [
    Patch(facecolor='orange', alpha=0.8, label='Retatrutide ▲'),
    Patch(facecolor='green', alpha=0.8, label='Patent Compounds ●')
]
ax.legend(handles=legend_elements, loc='upper left', fontsize=10, framealpha=0.9)

#ГРАФИК 3: GIPR vs GCGR 

ax = axes[2]

for idx, row in df_3d.iterrows():
    color, size, marker = get_color_and_size(row['id'])
    
    ax.scatter(row['abs_gipr'], row['abs_gcgr'],
              c=color, s=size, alpha=0.8,
              edgecolors='black', linewidth=2, marker=marker)
    
    ax.annotate(row['id'], 
               (row['abs_gipr'], row['abs_gcgr']),
               xytext=(9, -6), textcoords='offset points', 
               fontsize=9, fontweight='bold')

ax.set_xscale('log')
ax.set_yscale('log')

ax.set_xlabel('GIPR EC50, nM', 
             fontsize=12, fontweight='bold')
ax.set_ylabel('GCGR EC50, nM', 
             fontsize=12, fontweight='bold')

ax.set_title('ГРАФИК 3: GIPR × GCGR', 
            fontsize=13, fontweight='bold', pad=15)

ax.grid(True, alpha=0.3, which='both', linestyle=':')

add_reference_lines(ax, 'GIPR', 'GCGR', x_ref=1, y_ref=1)

legend_elements = [
    Patch(facecolor='orange', alpha=0.8, label='Retatrutide ▲'),
    Patch(facecolor='green', alpha=0.8, label='Patent Compounds ●')
]
ax.legend(handles=legend_elements, loc='upper left', fontsize=10, framealpha=0.9)


plt.suptitle('3 Взаимозависимых Графика: Абсолютные значения EC50\n в клеточных тестах активации GLP-1R × GIPR × GCGR',
            fontsize=16, fontweight='bold', y=1.02)

plt.tight_layout()
plt.show()
plt.close()
```

 Рассчитайте эти параметры для молекулы GPX-007 и приведите ответ в ячейке ниже. При использовании python приведите код, при использовании сторонних инструментов - кратко опишите алгоритм, по которому производили расчеты.

Количество дисульфидных связей, ароматических аминокислотных остатков и значение массы подсчитаны с помощью кода python (представлен ниже), А эмперическая формула была подсчитана вручную (с использованием количества каждой аминокислоты из кода)

```python
# Атомные массы согласно NIST
ATOMIC_MASSES = {
    'C': 12.011,
    'H': 1.00794,
    'N': 14.00674,
    'O': 15.9994,
    'S': 32.066,
    'H2O': 18.01528  
    # Отдельно записал воду, чтобы было проще расчитывать конденсацию
}

# Полный список протеиногенных аминокислот и 2 непротеиногенных
AMINO_ACIDS = {
    # Стандартные протеиногенные аминокислоты (20 штук)
    'A': {'name': 'Alanine', 'formula': 'C3H7NO2', 'mass': 89.09,
          'aromatic': False},
    'R': {'name': 'Arginine', 'formula': 'C6H14N4O2', 'mass': 174.20,
          'aromatic': False},
    'N': {'name': 'Asparagine', 'formula': 'C4H8N2O3', 'mass': 132.12,
          'aromatic': False},
    'D': {'name': 'Aspartic acid', 'formula': 'C4H7NO4', 'mass': 133.10,
          'aromatic': False},
    'C': {'name': 'Cysteine', 'formula': 'C3H7NO2S', 'mass': 121.16,
          'aromatic': False},
    'E': {'name': 'Glutamic acid', 'formula': 'C5H9NO4', 'mass': 147.13,
          'aromatic': False},
    'Q': {'name': 'Glutamine', 'formula': 'C5H10N2O3', 'mass': 146.15,
          'aromatic': False},
    'G': {'name': 'Glycine', 'formula': 'C2H5NO2', 'mass': 75.07,
          'aromatic': False},
    'H': {'name': 'Histidine', 'formula': 'C6H9N3O2', 'mass': 155.16,
          'aromatic': True},
    'I': {'name': 'Isoleucine', 'formula': 'C6H13NO2', 'mass': 131.17,
          'aromatic': False},
    'L': {'name': 'Leucine', 'formula': 'C6H13NO2', 'mass': 131.17,
          'aromatic': False},
    'K': {'name': 'Lysine', 'formula': 'C6H14N2O2', 'mass': 146.19,
          'aromatic': False},
    'M': {'name': 'Methionine', 'formula': 'C5H11NO2S', 'mass': 149.21,
          'aromatic': False},
    'F': {'name': 'Phenylalanine', 'formula': 'C9H11NO2', 'mass': 165.19,
          'aromatic': True},
    'P': {'name': 'Proline', 'formula': 'C5H9NO2', 'mass': 115.13,
          'aromatic': False},
    'S': {'name': 'Serine', 'formula': 'C3H7NO3', 'mass': 105.09,
          'aromatic': False},
    'T': {'name': 'Threonine', 'formula': 'C4H9NO3', 'mass': 119.12,
          'aromatic': False},
    'W': {'name': 'Tryptophan', 'formula': 'C11H12N2O2', 'mass': 204.23,
          'aromatic': True},
    'Y': {'name': 'Tyrosine', 'formula': 'C9H11NO3', 'mass': 181.19,
          'aromatic': True},
    'V': {'name': 'Valine', 'formula': 'C5H11NO2', 'mass': 117.15,
          'aromatic': False},
    
    # Непротеиногенные аминокислоты (используемые в GPX-007)
    '2meA': {'name': '2-Methylalanine', 'formula': 'C4H9NO2', 'mass': 103.12,
             'aromatic': False},
    '2meL': {'name': '2-Methylleucine', 'formula': 'C7H15NO2', 'mass': 145.20,
             'aromatic': False},
}

# Последовательность с модифицированными аминокислотами в квадратных скобках
GPX007_SEQUENCE_STR = "Y[2meA]QGTFTSDYSI[2meL]LDKKAAQAFIEYLLE[2meA]GPSSGAPPPS"


def parse_sequence(seq_str):
    """
    Парсит последовательность, выделяя модифицированные аминокислоты.
    Модифицированные аминокислоты указаны в квадратных скобках: [2meA]
    """
    parsed = []
    i = 0
    while i < len(seq_str):
        if seq_str[i] == '[':
            end = seq_str.index(']', i)
            aa = seq_str[i+1:end]
            parsed.append(aa)
            i = end + 1
        else:
            parsed.append(seq_str[i])
            i += 1
    return parsed


# Парсим последовательность (получаем список с аминокислотами)
gpx007_sequence = parse_sequence(GPX007_SEQUENCE_STR)

# ПОДСЧЕТ АМИНОКИСЛОТ И ИХ МАССЫ
# Подсчитываем количество каждой аминокислоты
aa_counts = Counter(gpx007_sequence)

composition_details = []
total_mass_aa = 0

for aa, count in sorted(aa_counts.items()):
    if aa in AMINO_ACIDS:
        aa_info = AMINO_ACIDS[aa]
        aa_mass = aa_info['mass']
        total_mass = aa_mass * count
        total_mass_aa += total_mass

print(f"ИТОГО: Суммарная масса всех 39 аминокислот = {total_mass_aa:.2f} Da")
print()

# Образование пептидных связей (конденсация)

# При образовании пептидной связи между двумя аминокислотами теряется H2O:
# R1-COOH + H2N-R2 → R1-CO-NH-R2 + H2O
# Количество пептидных связей = количество АК - 1

peptide_bonds = len(gpx007_sequence) - 1
water_loss = peptide_bonds * ATOMIC_MASSES['H2O']
condenced_mass = total_mass_aa - water_loss

# C-концевое амидирование

# C-конец амидирован: ...-S-CONH2 (вместо -COOH)
# Замена -OH на -NH2
# Потеря: OH (15.9994 + 1.00794 = 17.00734)
# Добавление: NH2 (14.00674 + 2×1.00794 = 16.02262)
# Разница: 16.02262 - 17.00734 = -0.98472 ≈ -0.98 Da

amidation_change = (14.00674 + 2 * 1.00794) - (15.9994 + 1.00794)
cond_Camid_mass = condenced_mass + amidation_change


# Шаг 5.5: Присоединение жирной кислоты C20


# Жирная кислота: C20H37O3, которая идёт от последнего атома азота в радикале (не от лизина)
# Весь ацильный радикал: C37H66O12N4
# Также при подсчёте массы надо будет учесть убыль одного атома водорода из аминогрупппы лизина
# В итоге нам надо прибавить к общему числу массу C37H65O12N4
acyl_rad_full_mass = 37 * ATOMIC_MASSES['C'] + 65 * ATOMIC_MASSES['H'] + 12 * ATOMIC_MASSES['O'] + 4 * ATOMIC_MASSES['N']

full_mass = cond_Camid_mass + acyl_rad_full_mass

empirical_formula = "C₂₃₀H₃₅₄N₄₉O₆₉"
# программу для автоматического подсчёта эмперической массы у меня не получилось написать
# поэтому подсчитал атомы самостоятельно

print(f"МОЛЕКУЛЯРНАЯ МАССА GPX-007:")
print(f"  Mw = {full_mass:.2f} Da")
print()

# Подсчет ароматических аминокислот
aromatic_count = 0
for aa, count in aa_counts.items():
    if AMINO_ACIDS[aa]['aromatic']:
        aromatic_count += count

# Подсчет непротеиногенных аминокислот
unnatural_count = 0
for aa, count in aa_counts.items():
    if aa in ['2meA', '2meL']:
        unnatural_count += count


# Подсчет цистеинов (для дисульфидных связей)
if aa_counts.get('C') is None:
    cysteine_count = 0
    disulfide_bonds = 0
else:
    cysteine_count = aa_counts.get('C')
    disulfide_bonds = cysteine_count // 2


# итоговая таблица
final_params = pd.DataFrame([
    {'Параметр': 'molecular_weight_Da', 'Значение': f'{full_mass:.2f}', 'Единицы': 'Da'},
    {'Параметр': 'empiric_formula', 'Значение': empirical_formula, 'Единицы': '-'},
    {'Параметр': 'aa_count', 'Значение': str(len(gpx007_sequence)), 'Единицы': 'остатков'},
    {'Параметр': 'aromatic_count', 'Значение': str(aromatic_count), 'Единицы': 'остатков'},
    {'Параметр': 'unnatural_count', 'Значение': str(unnatural_count), 'Единицы': 'остатков'},
    {'Параметр': 'disulfide_bonds', 'Значение': str(disulfide_bonds), 'Единицы': 'связей'}
])

print(final_params.to_string(index = False))
```
